Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Loading the dataset

```python
data = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")
print(data.head())
```

Checking for missing values

```python
print(data.isnull().sum())
```

Renaming the column names

```python
data.columns = ["States", "Date", "Frequency", "Estimated Unemployment Rate","Estimated Employed", "Estimated Labour Participation Rate","Region", "Longitude", "Latitude"]
```

Checking the correlation

```python
plt.style.use('seaborn-whitegrid')
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(data.corr(), ax=ax)
plt.show()
```

visualizing the data to analyze the unemployment rate

```python
plt.figure(figsize=(12, 10))
sns.histplot(x="Estimated Employed", hue="Region", data=data)
plt.title("Indian Unemployment")
plt.show()
```

unemployment rate according to different regions of India

```python
plt.figure(figsize=(12, 10))
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.title("Indian Unemployment")
plt.show()
```

create a dashboard to analyze the unemployment rate of each Indian state by region

```python
unemployment = data[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unemployment, path=["Region", "States"],
values="Estimated Unemployment Rate",
width=700, height=700, color_continuous_scale="RdYlGn",
title="Unemployment Rate in India")
figure.show()
```
